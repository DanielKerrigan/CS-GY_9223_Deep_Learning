```python
import rlcard
from rlcard.agents import RandomAgent
from rlcard.utils import set_global_seed
```

```python
# Make environment
env = rlcard.make('kuhn-poker', config={'seed': 0, 'record_action': True})
episode_num = 100
```

```python
# Set a global seed
set_global_seed(0)
```

```python
# Set up agents
agent_0 = RandomAgent(action_num=env.action_num)
agent_1 = RandomAgent(action_num=env.action_num)

env.set_agents([agent_0, agent_1])
```

```python
for episode in range(episode_num):

    # Generate data from the environment
    trajectories, _ = env.run(is_training=False)

    # Print out the trajectories
    print('\nEpisode {}\n'.format(episode))
    for ts in trajectories[0]:
        print(f'State: {ts[0]}, Action: {ts[1]}, Reward: {ts[2]}, Next State: {ts[3]}, Done: {ts[4]}')
    
    print('\nAction Record')
    if trajectories[0]:
        print(trajectories[0][-1][-2]['action_record'])
    else:
        print(trajectories[1][-1][-2]['action_record'])

    
    print('\n---------------')
```
